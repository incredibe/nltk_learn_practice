```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
quoteOfTheDay = "Wise men speak because they have something to say. Fools because they have to say something."
```

```python
sentencesOfQuote = sent_tokenize(quoteOfTheDay)
sentencesOfQuote
```

```python
wordsOfquote = word_tokenize(quoteOfTheDay)
wordsOfquote
```

We are about to filter words because it contain uncessary things as words like "." so on.
To Filter this we are using stopwords from nltk.corpus

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
random_str =  "When I interviewed the quarterback, he said they simply 'played a better game."
```

```python
wordsofRandomStr = word_tokenize(random_str)
wordsofRandomStr
```

```python
# creating empty list to store stopwords 
# stopwords are the words which we can ignore like  is, an, a, etc these words don't add a meaning
stop_words = set(stopwords.words('english'))
filteredList = [word for word in wordsofRandomStr if word.casefold() not in stop_words]
filteredList
```

```python
from nltk.stem import PorterStemmer
# creating stemmer 
stemmer = PorterStemmer()
```

```python
newStr = "The greatest glory in living lies not in never falling, but in rising every time we fall. If you look at what you have in life, you'll always have more. If you look at what you don't have in life, you'll never have enough. " 
```

```python
wordsOfnewStr = word_tokenize(newStr)
wordsOfnewStr
```

```python
# creating list of stemmer words from wordsOfnewStr
stemmedWords = [stemmer.stem(word) for word in wordsOfnewStr]
stemmedWords
```

```python
# since Porter stemming algorithm is not perfect and can sometimes produce incorrect stems for certain words. we can use snowstemmer
from nltk.stem import SnowballStemmer
snowStemmer = SnowballStemmer('english')
snowstemmedWords = [snowStemmer.stem(word) for word in wordsOfnewStr]
snowstemmedWords
```

```python
# to overcome the issues from the above stemming methods we are using lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatizedWords = [lemmatizer.lemmatize(word, po) for word in wordsOfnewStr]
lemmatizedWords
```
